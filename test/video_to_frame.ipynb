```python
##################    IMPORT    ########################
import os
import numpy as np
import cv2
import random
#########################################################
```

```python
##################    INIT    ########################
PROCESSED_DATA_PATH = "../dataset/sorted images/"
PROCESSED_DATA_FOLDER = os.listdir(PROCESSED_DATA_PATH)
SAVED_NUMPY_ARRAY_FOLDER ='../model/'

#########################################################
```

```python
##################  IMAGE PRE-PROCESSING   ########################


def create_training_data():
    CLASSES =  [0,1,2]
    TRAINING_DATA = []
    class_index = 0 

    #for each folders in the MNIST dataset, pre-process each image for training
    for folder in PROCESSED_DATA_FOLDER:

        #defining Path    
        IMAGE_PATH = os.path.join(PROCESSED_DATA_PATH, folder)   #defining the digit folder path
        CURRENT_FOLDER = os.listdir(IMAGE_PATH)           #defining wich folder to go in

        for image in CURRENT_FOLDER:

            #defining Path
            full_path = os.path.join(IMAGE_PATH  + "/" + image)   #defining the image full path
            #IMAGE_ARRAY.append(full_path)                                    #adding the full path to a list
            

            #image pre-processing
            CURRENT_CLASS = CLASSES[class_index]                         #defining the classes index
            image_loaded = cv2.imread(full_path)                         #Loading the image
            image_gray = cv2.cvtColor(image_loaded, cv2.COLOR_BGR2GRAY) #convert BlueGreenRed to Grayscale
            image_resized = cv2.resize(image_gray, (32,32))             #resizing the image 32x32 px
            TRAINING_DATA.append([image_resized, CURRENT_CLASS])          #append the resized image w/ corresponding index into TRAINING_DATA list

            print("folder ",folder, ": ", image, " has been processed - path: ",full_path ) 
       
        class_index += 1     #need this to iterate through our list "CLASSES"
    print("-----------------------------------------------------------")

    #splitting training data into features and label list
    X = []                                                  #our feature list, it will contain our handwritten digits
    y = []                                                  #our label list, it will contain our classes (0 to 9)
    random.shuffle(TRAINING_DATA)                           #we have to shuffle our data, because the for loop loaded them in order
    print("training data has been shuffle.")


    for features, label in TRAINING_DATA:
        X.append(features)
        y.append(label)
    print("training data split into features list (X), and label list (y).")
    print("X (before reshape): ", np.shape(X))

    #we can't pass a list to a Neural Network, so we have to convert our feature list, (X), in a numpy array
    X = np.array(X)
    X = X.reshape(-1, 32, 32, 1) # (-1 (here means "catch all images in the list"), width of the image, height of the image, color channel(1 is grayscale))

    #saving the generated X array for later use
    np.save(os.path.join(SAVED_NUMPY_ARRAY_FOLDER + "LineFollowing_features_v1.npy"), X) 
    np.save(os.path.join(SAVED_NUMPY_ARRAY_FOLDER + "LineFollowing_label_v1.npy"), y)
    print("X,y arrays saved at: ",SAVED_NUMPY_ARRAY_FOLDER)
    #X = np.load('features.npy') #loading


    # DEBUG
    print("-----------------------------------------------------------")
    print("training data shape: ",np.shape(TRAINING_DATA))
    print("X  (after reshape): ",np.shape(X), "    -   ", "y shape: ",np.shape(y), " - length of X: ", len(X))
    cv2.imshow("X head" ,X[0])
    print("y head: ", y[0])

    print("done.")

    cv2.waitKey(0)
    cv2.destroyAllWindows()
#########################################################


create_training_data()

```
